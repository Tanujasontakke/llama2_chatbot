Loading Model & Tokenizer
Preparing our session by loading both the llama model and its associated tokenizer.

The tokenizer will help in converting out text prompts into a format that the model can understand and process.

```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf" #meta-llama/Llama-2-7b-hf

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token = True)
```

Creating the llama Pipeline

setting up pipeline for text generation.
this pipeline simplifies the process of feeding prompts to our model and receiving generated text as output.

will take 2-3 mins to run

```python
from transformers import pipeline

llama_pipeline = pipeline(
    "text-generation", #LLM task
    model = model,
    torch_dtype = torch.float16,
    device_map = "auto",
)
```

Getting responses
now checking the response of llama to some sample queries.

```python
def get_llama_response(prompt: str)-> None:
  """
  Generate a response from the Llama model.

  Parameters:
    prompt(str): The user's input/question for the model.

  Returns:
    None: Prints the model's response.
  """
  sequences = llama_pipeline(
      prompt,
      do_sample = True,
      top_k = 10,
      num_return_sequences = 1,
      eos_token_id = tokenizer.eos_token_id,
      max_length = 256,
  )
  print("Chatbot:", sequences[0]['generated_text'])

prompt = 'I liked "Breaking Bad" and "Band of Brothers",Do you have any recommendations of other shows I might like?\n'
get_llama_response(prompt)

```

More Queries

```python
prompt = """
I am interested in AI and want to successfully build a chatbot, give me the steps that i should be doing to learn things and successfully build a bot?\n
"""
get_llama_response(prompt)
```
